# Scatter plots

Let's starts our journey with scatter plots in Matlibplot.

First of all let's ingest the data in and check that everything is fine.

```python
import pandas as pd

red_wines = pd.read_csv("../data/winequality-red.csv", sep=';', index_col='name')
white_wines = pd.read_csv("../data/winequality-white.csv", sep=';', index_col='name')
```

```python
red_wines.info()
```

```python
white_wines.info()
```

Let's now create a very single scatter plot. In the x-axis we will plot the fixed acidity of the wines and on the y axis we will plot the volatile acidity.

In addition, we will plt the red wines as red 'x's and the white wines as blue 'o's.

```python
import matplotlib.pyplot as plt

# This command is specific for Jupyter notebooks. It makes possible to visualise 
#the graphs within the notebook
# You need you run it at least one in the notebook before plotting with matplotlib
%matplotlib inline

# First of all we create a new figure with plt.figure()
# The figsize argument let you specify the size of the image (width, height)
plt.figure(figsize=(10,10))

# Plotting the red wines
# Use the parameters c (color) and marker for setting the right color and marker
plt.scatter(red_wines['fixed acidity'], red_wines['volatile acidity'], c='r', marker='x')
# Plotting the white wines
plt.scatter(white_wines['fixed acidity'], white_wines['volatile acidity'], c='b', marker='o')

# Setting tighter limits for the figure axes
# Computing the maximum and the minimum values
xmin = min(min(red_wines['fixed acidity']), min(white_wines['fixed acidity']))
xmax = max(max(red_wines['fixed acidity']), max(white_wines['fixed acidity']))
ymin = min(min(red_wines['volatile acidity']), min(white_wines['volatile acidity']))
ymax = max(max(red_wines['volatile acidity']), max(white_wines['volatile acidity']))
# Setting the right values
axes = plt.gca()
axes.set_xlim(xmin, xmax)
axes.set_ylim(ymin, ymax)
# Explore what happens if we don't use the set_xlim() & set_ylim() functions
# Comment out the two last code lines

# Set the legend using the legend() function
plt.legend(['Red wines', 'White wines'])

# Set the plot axes labels using xlabel() and ylabel()
plt.xlabel('Fixed acidity')
plt.ylabel('Volatile acidity')

# Set the plot title using title()
plt.title('Fixed acidity vs. volatile acidity')
```

Can you observe any difference in the distribution of the values of the red and the white wines?

## Plotting two graphs in one single figure

Let's try now to plot the distributions side by side in the same figure. For that, we will use the `subplot()` function.

```python
plt.figure(figsize=(20,10))

# We set the number of divisions and the number of graphs we want using the first two arguments
# The third argument let us choose the position where we want to plot:
# 1: one row
# 2: two columns
# 1: plotting in the first position of the matrix (first row, first column)
plt.subplot(1,2,1)

# Plotting the red wines
plt.scatter(red_wines['fixed acidity'], red_wines['volatile acidity'], c='r', marker='x')
plt.xlabel('Fixed acidity')
plt.ylabel('Volatile acidity')
plt.title('Fixed acidity vs. volatile acidity - Red wines')
plt.legend(['Red wines'])

# We move to the next plot using the subplot function again
# 1: one row
# 2: two columns
# 2: plotting in the second position of the matrix (first row, second column)
plt.subplot(1,2,2)

# Plotting the white wines
plt.scatter(white_wines['fixed acidity'], white_wines['volatile acidity'], c='b', marker='o')
plt.xlabel('Fixed acidity')
plt.ylabel('Volatile acidity')
plt.title('Fixed acidity vs. volatile acidity - White wines')
plt.legend(['White wines'])
```

Do you see any difference in the distributions?

Which figure is clearer?

Do you observe any issue with the figures axes? Let's solve the probem:

```python
plt.figure(figsize=(20,10))

# We set the number of divisions and the number of graphs we want using the first two arguments
# The third argument let us choose the position where we want to plot:
# 1: one row
# 2: two columns
# 1: plotting in the first position of the matrix (first row, first column)
plt.subplot(1,2,1)

# Plotting the red wines
plt.scatter(red_wines['fixed acidity'], red_wines['volatile acidity'], c='r', marker='x')
plt.xlabel('Fixed acidity')
plt.ylabel('Volatile acidity')
plt.title('Fixed acidity vs. volatile acidity - Red wines')
plt.legend(['Red wines'])
axes = plt.gca()
axes.set_xlim(xmin, xmax)
axes.set_ylim(ymin, ymax)

# We move to the next plot using the subplot function again
# 1: one row
# 2: two columns
# 2: plotting in the second position of the matrix (first row, second column)
plt.subplot(1,2,2)

# Plotting the white wines
plt.scatter(white_wines['fixed acidity'], white_wines['volatile acidity'], c='b', marker='o')
plt.xlabel('Fixed acidity')
plt.ylabel('Volatile acidity')
plt.title('Fixed acidity vs. volatile acidity - White wines')
plt.legend(['White wines'])
axes = plt.gca()
axes.set_xlim(xmin, xmax)
axes.set_ylim(ymin, ymax)
```

# Histograms

A histogram is a plot of the distribution of the values of a variable. You can use the `hist()` function for plotting histograms in Matplotlib.

Let's plot a histogram of the distribution of the wines density:

```python
plt.figure(figsize=(20,10))

# The first parameter of the function is the data we want to bin.
# We can use the 'bins' parameter to set the number of bins we want in our plots
plt.subplot(1,2,1)
plt.hist(red_wines['density'], bins=15, color='r')

plt.subplot(1,2,2)
plt.hist(white_wines['density'], bins=15, color='b')
```
